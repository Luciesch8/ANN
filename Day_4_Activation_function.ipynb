Importing library

```python
from numpy import array #For Array
from numpy import random # For Randomly
from numpy import dot # For doing dot product
from random import choice
```

Activation function

```python
import numpy as np

# step activation function
# activationFn = lambda x:0 if x<0 else 1 

# sigmoid activation function
#activationFn = lambda x: 1/(1+np.exp(-x))

# RELU activation function
#activationFn = lambda x: 0 if x<0 else x

# SOFTMAX activation function
activationFn = lambda x: np.exp(x) / np.sum(np.exp(x), axis=0)

error = []
```

Training dataset

```python
dataset = [
           (array([0,0,1]),0),  # array([x,y,b],e) x,y=Input , b=bias , e Expected O/P to validate
           (array([0,1,1]),1), 
           (array([1,0,1]),1), 
           (array([1,1,1]),1), 
]
print(dataset)
print(array([0,0,1]),0)
```

Initializing Random numbers for weights

```python
weights = random.rand(3)
weights
```

initializing additional variables

```python
r = 0.2 # learning rates
n = 100 # Number of Iteration
```

Training our model

```python
for j in range(n):
  x, expected = choice(dataset)
  #print(x,expected)
  result = dot(weights,x)
  err = expected-activationFn(result)
  error.append(err)
  weights += r * err * x # calculta the correction factor, added to weights to improve o/p
  # If the expected value is bigger, the weights shoud be increased
```
